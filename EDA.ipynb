```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load train data
train = pd.read_csv('train.csv')
train.tail()
```

```python
train.describe()
```

```python
dtype_df = train.dtypes.reset_index()
dtype_df.columns = ["Count", "Column Type"]
print(dtype_df.groupby("Column Type").aggregate('count').reset_index())
print("----------------------------")
print("We have only numerics values")
```

```python
# Check NA in train dataframe
print("Total Train Features with NaN Values = " + str(train.columns[train.isnull().sum() != 0].size))
if (train.columns[train.isnull().sum() != 0].size):
    print("Features with NaN => {}".format(list(train.columns[train.isnull().sum() != 0])))
    train[train.columns[train.isnull().sum() != 0]].isnull().sum().sort_values(ascending = False)
```

```python
plt.figure(figsize=(20,20))
plt.xticks( train.target, train.index.values ) # location, labels
plt.plot( train.target )
plt.xlabel("index")
plt.ylabel("value")
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(5,5))
plt.boxplot(train.target)
plt.show()
```

```python
# Bar Chart Frequency
fig = sns.distplot(train.target)
plt.xlabel("Target")
plt.ylabel("Frequency")
plt.show(fig)

def histogram(data, x_label, y_label, title):
    _, ax = plt.subplots()
    ax.hist(data, color = '#539caf')
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)

# Call the function to create plot
histogram(data = train.target
           , x_label = 'Target'
           , y_label = 'Frequency'
           , title = 'Distribution of Target')
```

```python
# Scatter plot short by target values
plt.figure(figsize=(8,6))
plt.scatter(range(train.shape[0]), np.sort(train['target'].values))
plt.xlabel('index', fontsize=12)
plt.ylabel('Target', fontsize=12)
plt.title("Target Distribution", fontsize=14)
plt.show()
```
